```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.firefox.options import Options
from bs4 import BeautifulSoup
```

```python
import time
```

```python
# Function to scrape data from Dicoding
def scraper(url):
    try:
        # Configure WebDriver to use headless Firefox
        options = Options()
        options.add_argument('-headless')
        driver = webdriver.Firefox(options=options)

        # Get the URL given
        driver.get(url)

        # BeautifulSoup will parse the URL
        content = driver.page_source
        soup = BeautifulSoup(content, 'html.parser')

        # Tambahkan waktu tunggu setelah memuat halaman
        driver.get(url)
        time.sleep(5)  # Tunggu 5 detik agar elemen dimuat sepenuhnya

        WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, 'ipc-rating-star--voteCount')))
        # Prepare the variable for storing film data
        film_list = []

        # BeautifulSoup will find the CSS class that contains product container
        for film in soup.find_all('li', class_='ipc-metadata-list-summary-item'):
            
            vote_count_element = film.find('span', class_='ipc-rating-star--voteCount')

            if vote_count_element:
                jumlah_perating = vote_count_element.text
                print(f"Jumlah perating: {jumlah_perating}")
            else:
                print("Vote count not found")

            # Get the text from the specified element and assign them to the variables
            judul_film = film.find('h3', class_='ipc-title__text').text if film.find('h3', class_='ipc-title__text') else 'N/A'
            tahun_film = film.find_all('span', {'class':'sc-b189961a-8 hCbzGp dli-title-metadata-item'})[0].text if len(film.find_all('span', {'class':'sc-b189961a-8 hCbzGp dli-title-metadata-item'})) > 0 else 'N/A'
            durasi_film = film.find_all('span', {'class':'sc-b189961a-8 hCbzGp dli-title-metadata-item'})[1].text if len(film.find_all('span', {'class':'sc-b189961a-8 hCbzGp dli-title-metadata-item'})) > 1 else 'N/A'
            batas_usia = film.find_all('span', {'class':'sc-b189961a-8 hCbzGp dli-title-metadata-item'})[2].text if len(film.find_all('span', {'class':'sc-b189961a-8 hCbzGp dli-title-metadata-item'})) > 2 else 'N/A'
            sinopsis_film = film.select('div.ipc-html-content-inner-div')[0].text if film.select('div.ipc-html-content-inner-div') else 'N/A'
            rating_film = film.find('span', class_='ipc-rating-star--rating').text.replace('\xa0', '')
            jumlah_perating = film.find('span', class_='ipc-rating-star--voteCount').text.replace('\xa0', '')
           
            
            # Append the film data to the list
            film_list.append(
                {
                    'judul film': judul_film,
                    'tahun film': tahun_film,
                    'durasi film': durasi_film,
                    'batas usia': batas_usia,
                    'sinopsis film': sinopsis_film,
                    'rating film': rating_film,
                    'jumlah perating': jumlah_perating,
                }
            )

        # Close the WebDriver
        driver.quit()
 
        return film_list
    
    except Exception as e:
        # Print the error message
        print('An error occurred: ', e)
 
        # Close the WebDriver
        driver.quit()

# Example call to the function
film_data = scraper("https://www.imdb.com/search/title/?groups=top_1000&count=200&sort=user_rating,asc")
print(film_data)
```
